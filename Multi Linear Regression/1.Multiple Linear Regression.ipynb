```python
import pandas as pd
dataset=pd.read_csv("50_Startups.csv")
```

```python
dataset.head(5)
```

```python
# Convert categorical value into numeric value
dataset = pd.get_dummies(dataset, dtype=int, drop_first=True)
```

```python
dataset.head(5)
```

```python
dataset.columns
```

```python
independent = dataset[['R&D Spend', 'Administration', 'Marketing Spend','State_Florida', 'State_New York']]
independent.head(5)
```

```python
dependent=dataset[['Profit']]
dependent.head(5)
```

```python
# importing train_test_split library from sklearn.model_selection library collection
from sklearn.model_selection import train_test_split
# spliting the dataset to train and test with ratio of 70:30 and assign to variable X_train, X_test, y_train and y_test. 
#for industry standard X represents input, y represents output
X_train,X_test,y_train,y_test=train_test_split(independent, dependent,test_size=0.30, random_state=0)
```

```python
# importing LinearRegression library from sklearn.linear_model library collection
from sklearn.linear_model import LinearRegression
#assign LinearRegression to a regressor variable
regressor = LinearRegression()
# Train the model using fit method and passing both input and output training value 
regressor.fit(X_train,y_train)
```

```python
# coef_ variable contains weight of the Linear regression
weight = regressor.coef_
weight
```

```python
# intercept_ variable contains the bias value or starting value or intercept value
bias = regressor.intercept_
bias
```

```python
# predict the value passsing test input value
y_pred = regressor.predict(X_test)
```

```python
# Model evaluation using R2_score. Always output between 0 to 1. If the value is near to 1 then the model good model.
# importing r2_score library from sklearn.metrics collections
from sklearn.metrics import r2_score
# evaluate the model using original test output value  vs predicted test value and assign to variable
r_score = r2_score(y_test, y_pred)
```

```python
# diplay variable
r_score
```

```python
# Save the model
# import pickle library
import pickle
# Assign the model filename into a filename variable
filename = "finalized_model_mlr.sav"
```

```python
# Dump the model / save the model into filename
pickle.dump(regressor,open(filename,'wb'))
```

```python
#Verifying the saved model. 
#Load the saved model from the saved file
loaded_model = pickle.load(open("finalized_model_mlr.sav",'rb'))
```
