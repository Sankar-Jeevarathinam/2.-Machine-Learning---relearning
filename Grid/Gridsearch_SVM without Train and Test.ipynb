```python
import pandas as pd
dataset = pd.read_csv("insurance_pre.csv")
```

```python
dataset.head(2)
```

```python
dataset = pd.get_dummies(dataset, dtype=int, drop_first=True)
```

```python
dataset.head(2)
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children',  'sex_male', 'smoker_yes']]
dependent=dataset[['charges']]
```

```python
#from sklearn.model_selection import train_test_split
#X_train, X_test, y_train, y_test = train_test_split(independent, dependent, test_size=0.3, random_state=0)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
param_grid = {'kernel':['rbf','sigmoid','linear'], 
              'C':[10,100,1000,2000,3000], 'gamma':['auto','scale']}
grid = GridSearchCV(SVR(), param_grid, refit=True, verbose=3, n_jobs=-1)
grid.fit(independent, dependent)
```

```python
result=grid.cv_results_
print(result)
```

```python
#from sklearn.metrics import r2_score
#r_score = r2_score(y_test,grid_pred)
#print("The r_score value of best parameter {}:".format(grid.best_params_), r_score)
print("The best parameter is {}:".format(grid.best_params_))
```
