```python
import pandas as pd
dataset=pd.read_csv("50_Startups.csv")
```

```python
# Convert categorical value into numeric value
dataset = pd.get_dummies(dataset, dtype=int, drop_first=True)
```

```python
dataset.columns
```

```python
independent = dataset[['R&D Spend', 'Administration', 'Marketing Spend','State_Florida', 'State_New York']]
dependent=dataset[['Profit']]
```

```python
# importing train_test_split library from sklearn.model_selection library collection
# spliting the dataset to train and test with ratio of 70:30 and assign to variable X_train, X_test, y_train and y_test. 
#for industry standard X represents input, y represents output

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent, dependent,test_size=0.30, random_state=0)
```

```python
# importing SVR library from sklearn.svm library collection
from sklearn.svm import SVR
regressor = SVR(kernel="linear")
# Train the model using fit method and passing both input and output training value 
regressor.fit(X_train,y_train)
```

```python
regressor.coef_
```

```python
regressor.n_support_
```

```python
regressor.support_
```

```python
regressor.intercept_
```

```python
y_pred = regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r_score = r2_score(y_test, y_pred)
r_score
```

```python
regressor = SVR(kernel="rbf")
regressor.fit(X_train,y_train)
y_pred = regressor.predict(X_test)
r_score = r2_score(y_test, y_pred)
r_score
```

```python
regressor = SVR(kernel="poly")
regressor.fit(X_train,y_train)
y_pred = regressor.predict(X_test)
r_score = r2_score(y_test, y_pred)
r_score
```
